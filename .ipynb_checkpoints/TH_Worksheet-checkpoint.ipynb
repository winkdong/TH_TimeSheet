### C1 C2指定项目的总工时

```python
# df1.loc[lambda df: df.A > 0, :]
project = data.loc[lambda df: df["项目名称"]=="和田昆仑湖公园"]
pt_s = project["标准工时"].sum()
pt_o = project["加班工时"].sum()
print(pt_s)
print(pt_o)
print(pt_s+pt_o)
```

### C1 B1指定项目的总工时及参与个人所占工时

```python
c1b1 = project.groupby(["姓名"])["标准工时","加班工时"].sum()
c1b1
```

```python
c1b1_worktime = list(c1b1["标准工时"])
c1b1_overwork = list(c1b1["加班工时"])
N = len(c1b1_worktime)
ind = np.arange(N)    # the x locations for the groups
width = 0.5           # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, c1b1_worktime, width, color='r')
p2 = plt.bar(ind, c1b1_overwork, width, color='y',bottom=c1b1_worktime)
plt.xticks(range(N), c1b1.index.values)
plt.show()
```

### C2 B2 指定项目的总加班工时及参与个人加班工时

```python
c2b2 = project.groupby(["姓名"])["加班工时"].sum()
c2b2
```

```python
labels = c2b2.index.values
sizes = list(c2b2)

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels,autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
